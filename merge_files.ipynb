```python
import os

# Path to the folder containing the .txt files
folder_path = "/content/drive/MyDrive/New Word Prediction LSTM/eminem_lyrics"

# Initialize an empty string to store the merged content
merged_content = ""

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r", encoding="utf-8") as file:
            file_content = file.read()
            merged_content += file_content + "\n"

# Write the merged content to a new file
with open("/content/drive/MyDrive/New Word Prediction LSTM/dataset.txt", "w", encoding="utf-8") as dataset:
    dataset.write(merged_content)

print("Files merged successfully.")
```
